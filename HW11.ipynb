```python
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# files.upload() # used for uploading img to Colab
```

## Original Image

```python
x = np.loadtxt("clownImage.txt")

plt.gray()
plt.imshow(x)
plt.show()
```

## SVD

```python
u, s, v = np.linalg.svd(x, full_matrices=True)
```

## Rank Approximations

```python
for rank in [5,15,25]:
  svd_decomp = np.matrix(u[:, :rank]) * np.diag(s[:rank]) * np.matrix(v[:rank, :])

  plt.gray()
  plt.title(f"k={rank}")
  plt.imshow(svd_decomp)
  plt.show()
```
